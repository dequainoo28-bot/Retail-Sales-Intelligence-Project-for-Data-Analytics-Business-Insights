```python
# We now train the model.
# We define the features.
X = pd.DataFrame(np.c_[numeric_df['Profit'],numeric_df['Unit Price']], columns = ['Profit','Unit Price'])
# We now print the data we have inspected to train
print(X.head())
```

```python
# define the target (So the x data becomes what we will use to find the y that is the target)
y = numeric_df['Revenue']
print(y.head())
```

```python
# We now create training and test data.
# We first set or define our test and training data.
from sklearn.model_selection import train_test_split
X_test,X_train, y_test,y_train, = train_test_split(X, y, test_size = 0.25, random_state =5)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
# We now create the model 
from sklearn.linear_model import LinearRegression
sale_model = LinearRegression()
sale_model.fit(X_train, y_train)
```

```python
# we do know that the formula for linear regression is y= mx+c( and the x can be many )
# We will now print out the m and c that is the gradient, and the intercept.
print(sale_model.coef_)
print (sale_model.intercept_)
```

```python
# We now calculate it as the formula we know.
# We will now insect our data.
sale_pred = sale_model.predict(X_test)
print('R-squared: %.4f' % sale_model.score(X_test, y_test))
```

```python
# visualize error
plt.scatter(x = y_test, y = sale_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.show()
```

```python
# find the mean Squared Error MSE.
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, sale_pred)
print(MSE)
```
